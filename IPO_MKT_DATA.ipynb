```python
tickers = ['FC', 'LNKD', 'GRPN', 'SNAP', 'TWTR', 'BABA', 'ETSY', 'FIT', 'WDAY', 'GPRO', 'APRN', 'RACE', 'GM', 'SHAK', 'SFIX', 'TSLA']
```

```python
prices_df = pd.DataFrame()
for ticker in tickers:
        temp_df = data.DataReader(str(ticker), 'morningstar', '2012-01-01', '2018-05-23')
        prices_df = prices_df.append(temp_df, ignore_index = True)
```

```python
prices_df.info()
```

```python
_newprices_df = pd.DataFrame()
for ticker in tickers:
        temp_df = data.DataReader(str(ticker), 'morningstar', '2010-01-01', '2018-05-24')
        _newprices_df = _newprices_df.append(temp_df)
```
